```python
import pandas as pd
```

```python
df = pd.read_csv('../data/pew.csv')
df.head()
```

```python
df_long = pd.melt(df, id_vars = ['religion'], var_name = 'income', value_name = 'count' )
df_long.head()
```

```python
billboard = pd.read_csv('../data/billboard.csv')
billboard_melt = pd.melt(billboard, id_vars = ['year','artist','track','time','date.entered'], var_name = 'week', value_name = 'rating')
billboard_melt.head()
```

```python
ebola = pd.read_csv('../data/country_timeseries.csv')
ebola_melt = pd.melt(ebola, id_vars = ['Date', 'Day'], var_name = 'variable', value_name = 'count')
ebola_melt.head()

ebola_split = ebola_melt.variable.str.split("_")

ebola_melt['stats'] = ebola_split[0][0]
ebola_melt['country'] = ebola_split[0][1]

ebola_melt[['stats_e', 'country_e']] = ebola_melt.variable.str.split('_', expand = True)
ebola_melt.head()
```

```python
weather = pd.read_csv('../data/weather.csv')

weather.head()
weather_melt = pd.melt(weather, id_vars = ['id','year','month','element'], var_name = 'days', value_name = 'temp')
weather_melt.head()

weather_tidy = weather_melt.pivot_table(index = ['id','year','month','days'],columns='element', values='temp')
weather_tidy.reset_index()
```
